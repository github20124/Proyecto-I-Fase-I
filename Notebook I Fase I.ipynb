```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Carga de Datos

```python
dataopen = pd.read_csv("openpowerlifting.csv")
dataopen.head()
```

# Visualización con Gráficos

```python
# Distribución de Genero en Histograma
plt.figure(figsize=(7, 5))
sns.histplot(dataopen['Sex'].dropna(), bins=10, kde=True)
plt.title('Distribución de Género')
plt.xlabel('Género')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Distribucion de Año en histograma
plt.figure(figsize=(10, 6))
sns.histplot(dataopen['Age'].dropna(), bins=30, kde=True)
plt.title('Distribución de Edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Distribución de Equipamiento en Gráfico de Pie
equipment_counts = dataopen['Equipment'].value_counts()
plt.figure(figsize=(10, 6))
plt.pie(equipment_counts, labels=equipment_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proporción del Tipo de Equipamiento')
plt.show()
```

```python
# Densidad del peso Corporal en gráfico de densidad
plt.figure(figsize=(10, 6))
sns.kdeplot(dataopen['BodyweightKg'].dropna(), shade=True)
plt.title('Distribución de la Densidad del Peso Corporal')
plt.xlabel('Peso Corporal (Kg)')
plt.ylabel('Densidad')
plt.show()
```

```python
# Distribution of Best Squat en histograma
plt.figure(figsize=(10, 6))
sns.histplot(dataopen['BestSquatKg'].dropna(), bins=30, kde=True)
plt.title('Distribución de la Mejor Marca en Sentadilla')
plt.xlabel('Mejor Sentadilla (Kg)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
sns.pairplot(dataopen[['BestBenchKg']].dropna())
plt.suptitle('Gráfico de Pares', y=1.02)
plt.show()
```

```python
# Distribución de Best Deadlift
plt.figure(figsize=(10, 6))
sns.histplot(dataopen['BestDeadliftKg'].dropna(), bins=30, kde=True)
plt.title('Distribución de la Mejor Marca en Peso Muerto')
plt.xlabel('Mejor Peso Muerto (Kg)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Distribución de Total Kg
plt.figure(figsize=(10, 6))
sns.histplot(dataopen['TotalKg'].dropna(), bins=30, kde=True)
plt.title('Distribución del Total Levantado')
plt.xlabel('Total Levantado (Kg)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.figure(figsize=(12,7))
dataopen[dataopen].value_counts(normalize=True).sort_values(ascending=False).plot.bar()
plt.xlabel("Puntaje Wilks")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Distribución de Wilks Score
plt.figure(figsize=(10, 6))
sns.histplot(dataopen['Wilks'].dropna(), bins=30, kde=True)
plt.title('Distribución del Puntaje Wilks')
plt.xlabel('Puntaje Wilks')
plt.ylabel('Frecuencia')
plt.show()
```
